```python
# From above i see that AMH and II beta-HCG are stored as strings
# Convert AMH and II beta-HCG to numeric
col = ["AMH_(ng/mL)", "II_beta-HCG_(mIU/mL)"]
df[col] = df[col].apply(pd.to_numeric, errors ="coerce" , axis = 1)

# Fill na as 0 for continuous var II_beta-HCG_(mIU/mL) and AMH
df["II_beta-HCG_(mIU/mL)"].fillna(0.0, inplace = True)
df["AMH_(ng/mL)"].fillna(0.0, inplace = True)


# Some Y/N categorical variables appear as float. Convert them to int.
df["Fast_food_(Y/N)"] = df["Fast_food_(Y/N)"].astype("Int64") #handles 

# Blood group label is confusing. Change int labeling back to blood group name
# We can use onehot encoding in the modeling step for blood group
blood_groups = {11: "A+", 
                12: "A-",
                13: "B+",
                14: "B-",
                15: "O+",
                16: "O-",
                17: "AB+",
                18: "AB-"}
df["Blood_Group"].replace(blood_groups, inplace = True)

df["Cycle_(R/I)"].value_counts()
```

### Modeling

#### 1. Random Forest

```python
seed = 50  # so that the result is reproducible # How to set seed - like which number to choose?

# Import libraries for modeling
from sklearn.model_selection import train_test_split # split data into train and test
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.ensemble import RandomForestClassifier # randomforest model
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, roc_auc_score, roc_curve, f1_score
from sklearn.metrics import classification_report
from sklearn.pipeline import make_pipeline
```

```python
# Create feature and truth dataframes
pcos_df = df.copy()
y = df.pop("PCOS_(Y/N)") # truth
X = df.drop("Patient_File_No", axis = 1) # feature df. removing patient id since it is not a feature
df.isnull().sum()
```

```python
# Create a training and validation set
# setting a seed for random state ensure reproducible split
# stratify argument ensure proportion of PCOS Y and N is similar in the training and validation sets
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.333, random_state = seed, stratify=y )

lis = [X_train, X_test, y_train, y_test]
for each in lis:
    each.shape
```

```python
# Encoding categorical variables

# create a list of columns that are categorical variables and need encoding
# leaving categorical variables that are already binary as is
features_to_encode = ["Blood_Group", "Cycle_(R/I)"]

# create constructor to handle categorical column with multiple categories
# constructor takes the categorical variables and transforms them into a correct type; here uses OneHotEncoding
# only transforms features_to_encode and leaves other columns as is in output
col_trans = make_column_transformer(
                        (OneHotEncoder(),features_to_encode),
                        remainder = "passthrough"
                        )
```

```python
# create RF  classifier
rfc = RandomForestClassifier(random_state=seed, 
                             bootstrap = True, 
                             oob_score = True,
                             max_features = "auto")
```

```python
# combine constructor and classifier into a pipeline
# Classifier received inputs transformed by constructor
pipe = make_pipeline(col_trans, rfc)
pipe.fit(X_train, y_train)
```

```python
# Making prediction on the test set
y_pred = pipe.predict(X_test)
```

```python
acc = round(accuracy_score(y_test,y_pred),3)*100

print("The accuracy of the model is {a:.2f} %".format(a=acc))
```

```python
# Make probability predictions
train_probs = pipe.predict_proba(X_train)[:,1]
train_predictions = pipe.predict(X_train)

test_probs = pipe.predict_proba(X_test)[:, 1]
```

```python
# Calculate AUC scores
train_auc_score = roc_auc_score(y_train, train_probs)
test_auc_score = roc_auc_score(y_test, test_probs)

# Print statements
train_roc = 'Train ROC AUC Score: {score:.2f}'
test_roc = "Test ROC AUC Score: {score:.2f}"

print(train_roc.format(score = train_auc_score))
print(test_roc.format(score = test_auc_score))
```
